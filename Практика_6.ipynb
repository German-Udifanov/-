```python
import numpy as np
import matplotlib.pyplot as plt
# Генерация случайных величин x
x=np.random.uniform(low=-5,high=7,size=50)
# Вычисление значений функции f(x)
y_true=(1/3)*x* np.cos(x)
# Добавление случайной компоненты E
E = np.random.uniform(low=-1.5, high=1.5, size=50)
y_true += E
# Построение графиков
plt.plot(x, y_true, 'bo', label='Исходная функция')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
